### Execise 1b)
Write a program that solves the MathTechQuest problem.

The program should consists of three functions:
- check_combinations: check all possible combinations of the digits and returns the valid combinations
- print_combinations: prints all valid combinations to the problem
- main: executes the program and prompts the user if all solutions to the problem should be printed


```python
def check_combinations():
    digs = [str(i) for i in range(10)]
    all_perm = permutations(digs)
    valid_c = []

    for p in all_perm:
        a, b, c, d, e, f, g, h, i, j = p

        abc = int(a + b + c)
        de = int(d + e)
        fgh = int(f + g + h)
        ij = int(i + j)

        if abc*de - fgh*ij == 0:
            valid_c.append(p)
    return(valid_c)
```

```python
def print_combinations():
    for valid_c in check_combinations():
        print(valid_c)
```

```python
def main():
    print('this programs checks how which combinations of numbers from 0-9 satify the equation (abc*de)-(fgh*ij) = 0')
    check_combinations() 
    if len(check_combinations()) == 198:
        print('there are 198 solutions')
        see_p = input('do you want to see the soluttions (type "yes" or "no"): ')
        if see_p == 'yes':
            print_combinations()

        
```

Modify the random password generator from last week by replacing the for loop and randint to choose random characters with the methods choice, choices and shuffle from random.

Create a function called generate_password that creates a random password consisting of letters and a single digit and symbol:
- The function takes one input: the length of the password
- The function should return the generated password as a string.
    
    Note that the generated password should be at least 4 characters long, and it should not start with a symbol or digit.

Demonstrate the function by generating a random password of 10 characters.
- Hint: use the string module (see here) to generate sequences of letters, digits and symbols to choose from

```python
import string as st
import random as rd

def generate_password():
    length = int(input('write the length of the password: '))

    pw = rd.choices(st.ascii_letters, k = length-2)
    pw = ''.join(pw)
    pw_digit = rd.choice(st.digits)
    pw_symbol = rd.choice(st.punctuation)
    password = list(pw + pw_digit + pw_symbol)
    rd.shuffle(password)
    password[0] = str(rd.choice(st.ascii_letters))
    final_password = ''.join(password)

    return(final_password)
```
